Rules for 21

Take the Joker out of the pack.

Start with 2 cards. 

you can either stick or twist. Stick means you don't ask for another card. 

Twist means you ask for another card.

Keep adding cards until they total 21 or as high as you can get.

Ace = 1 or 11

Court Cards are 10

A player can have a maximum of 5 cards. If a player has 21 or less in 5 cards he wins the game if 21 or under in 2, 3, or 4 cards - its a 5 card trick.

The player that has the highest amount up to 21 wins.

If a player gets higher than 21 he loses and has to say Bust.



Import libraries

```
import numpy as np
import pandas as pd
from random import randint
from numpy import array
```

Initialise

```
games = 1000
round = 0
winner = []
game_no = 0
game = 0
```

Game

```
for game in range(games):
  deal1 = 0
  deal2 = 0
  score1 = 0
  score2 = 0
  cards1 = []
  cards2 = []
  win = 0
  game_no = game_no + 1
  while (score1 < 21) and (score2 < 21):
    round = round + 1
    deal1 = randint(1,10)
    if deal1 == 1 and score1 < 10:
      deal1 = 11
    cards1.append(deal1)
    score1 = sum(cards1)
    print("round: ", round, "deal1: ", deal1, "cards1: ", cards1, " score1: ", score1)
    deal2 = randint(1,10)
    if deal2 == 1 and score2 < 10:
      deal2 = 11
    cards2.append(deal2)
    score2 = sum(cards2)
    print("round: ", round, "deal2: ", deal2, "cards2: ", cards2, " score2: ", score2)
  if score1 == score2:
    win = 0
  elif score1 > 21 and score2 > 21:
    win = 0
  elif score1 > score2 and score1 <= 21 or score2 > 21:
    win = 1
  elif score2 > score1 and score2 <= 21 or score1 >21:
    win = 2
  print("The winner is: ", win)
  winner.append(win)
print("The sequence of winners is: ", winner)
  
```
