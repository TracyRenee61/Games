Roman numerals are a number system that originated in ancient Rome and remained the usual way of writing numbers throughout Europe well into the late middel ages. 
Modern usage employs seven symbols:-

I = 1

V = 5

X = 10

L = 50

C = 100

D = 500

M = 1000

Roman numerals are a base-10 number system but instead of place value notations they use symbols to represent fixed values.

My converter

```
num_dic = {'M': 1000, 'D': 500, 'C': 100, 'L': 50, 'X': 10, 'V': 5, 'I': 1}

def roman2arabic(numeral):
    answer = 0
    n = len(numeral)
    for (idx, c) in enumerate(numeral):
      if idx < n-1 and num_dic[c] < num_dic[numeral[idx+1]]:
        answer = answer - num_dic[c]
      else:
        answer = answer + num_dic[c]
    return answer
```
